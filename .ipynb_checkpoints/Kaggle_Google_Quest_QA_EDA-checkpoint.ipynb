```python
import re
```

```python
def basic_cleaning(text):
    pat1 = r"(\s{2,})|(\\n)"
    pat2 = r"\\"
    pat3 = r"\s$"
    edit = re.sub(pat1, ' ', text)
    edit2 = re.sub(pat2, '', edit)
    edit3 = re.sub(pat3, '', edit2)
    return edit3
```
